# SILVER DATA LOAD

```python
import psycopg2
```

```python
# DATAWAREHOUSE CONFIGURATION
DB_CONFIG = {
    "host": "localhost",        # Postgres host
    "port": 5432,               # default port
    "dbname": "DataWarehouse",            # Database
    "user": "userhere",          # DB user
    "password": "passwordhere"       # DB user password
}
```

```python
conn = psycopg2.connect(**DB_CONFIG)
conn.autocommit = True  # REQUIRED for CALL + TRUNCATE

try:
    with conn.cursor() as cur:
        cur.execute("CALL silver.load_silver();")

        # Print RAISE NOTICE output
        print("---- Procedure notices ----")
        for notice in conn.notices:
            print(notice.strip())

    print("Silver load completed successfully.")

except Exception as e:
    print("Error while executing procedure:", e)

finally:
    conn.close()
```
